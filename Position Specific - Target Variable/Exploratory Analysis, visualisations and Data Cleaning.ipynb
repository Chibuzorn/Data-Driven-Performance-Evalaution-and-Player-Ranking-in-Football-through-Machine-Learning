```python
# Calculating the overall rating for each player by averaging their ratings, which the who scored ratings
overall_ratings = football_data.groupby('player')['original_rating'].mean().sort_values(ascending=False)

# Getting the top 5 players with the highest overall ratings
top_5_overall_players = overall_ratings.head(5)

# Displaying the top 5 players with their overall ratings
top_5_overall_players
```

```python
# top played based on players that have played minimum of 10 matches in the whole competition
# Setting the minimum number of matches threshold
min_matches = 10

# Filtering players who have played at least the minimum number of matches using the 'date' column
players_with_min_matches = football_data.groupby('player').filter(lambda x: x['date'].nunique() >= min_matches)

# Calculating the average rating for these players
average_ratings_min_matches = players_with_min_matches.groupby('player')['original_rating'].mean().sort_values(ascending=False)

# Getting the top 5 players with minimum matches threshold
top_5_players_min_matches = average_ratings_min_matches.head(5)

# Calculating the number of matches each player played using the 'date' column
matches_played_by_date = players_with_min_matches.groupby('player')['date'].nunique()

# Create DataFrame for top 5 players with their average ratings and number of matches
top_5_df = pd.DataFrame({
    'average_rating': top_5_players_min_matches,
    'matches_played': matches_played_by_date[top_5_players_min_matches.index]
})

# Print the combined results
print("Top Players Analysis (with Number of Matches)")
print(top_5_df)

# Visualize the results
fig, ax = plt.subplots(figsize=(10, 6))

# Plot for Top Players (Min Matches >= 10)
ax.barh(top_5_df.index, top_5_df['average_rating'], color='blue')
for i, (rating, matches) in enumerate(zip(top_5_df['average_rating'], top_5_df['matches_played'])):
    ax.text(rating, i, f'{matches} matches', va='center', ha='left', color='black', fontsize=12)
ax.set_title('Top Players (Min Matches >= 10) - Using Date')
ax.set_xlabel('Average Rating')
ax.set_ylabel('Player')

plt.tight_layout()
plt.show()
```

```python
#top players with minimum of 10 matches in each competitons
# Setting the minimum number of matches threshold
min_matches = 10

# Get the unique competitions
competitions = football_data['competition'].unique()

# Iterate through each competition
for competition in competitions:
    # Filter data for the current competition
    competition_data = football_data[football_data['competition'] == competition]
    
    # Filtering players who have played at least the minimum number of matches using the 'date' column
    players_with_min_matches = competition_data.groupby('player').filter(lambda x: x['date'].nunique() >= min_matches)
    
    if not players_with_min_matches.empty:
        # Calculating the average rating for these players
        average_ratings_min_matches = players_with_min_matches.groupby('player')['original_rating'].mean().sort_values(ascending=False)
        
        # Getting the top 5 players with minimum matches threshold
        top_5_players_min_matches = average_ratings_min_matches.head(5)
        
        # Calculating the number of matches each player played using the 'date' column
        matches_played_by_date = players_with_min_matches.groupby('player')['date'].nunique()
        
        # Create DataFrame for top 5 players with their average ratings and number of matches
        top_5_df = pd.DataFrame({
            'average_rating': top_5_players_min_matches,
            'matches_played': matches_played_by_date[top_5_players_min_matches.index]
        })
        
        # Print the combined results
        print(f"Top Players Analysis in {competition} (with Number of Matches)")
        print(top_5_df)
        
        # Visualize the results
        fig, ax = plt.subplots(figsize=(10, 6))
        
        # Plot for Top Players (Min Matches >= 10)
        ax.barh(top_5_df.index, top_5_df['average_rating'], color='blue')
        for i, (rating, matches) in enumerate(zip(top_5_df['average_rating'], top_5_df['matches_played'])):
            ax.text(rating, i, f'{matches} matches', va='center', ha='left', color='black', fontsize=12)
        ax.set_title(f'Top Players in {competition} (Min Matches >= 10) - Using Date')
        ax.set_xlabel('Average Rating')
        ax.set_ylabel('Player')
        
        plt.tight_layout()
        plt.show()
```

## To what extent are the ratings assigned to the players' performances biased by the final outcome of the match?
Analyze the correlation between match outcomes (win, loss) and player ratings.

## Approach:
Correlation analysis and visualization of ratings versus match outcomes.

```python
football_data['outcome'] = football_data['win'] - football_data['lost']
sns.boxplot(x='outcome', y='original_rating', data=football_data)
plt.title('Player Ratings vs Match Outcome')
plt.xlabel('Match Outcome (Win: 1, Draw: 0, Loss: -1)')
plt.ylabel('Player Rating')
plt.show()

# Statistical test to check for bias
from scipy.stats import ttest_ind

win_ratings = football_data[football_data['outcome'] == 1]['original_rating']
loss_ratings = football_data[football_data['outcome'] == -1]['original_rating']
t_stat, p_value = ttest_ind(win_ratings, loss_ratings)
print(f'T-test: t-statistic={t_stat}, p-value={p_value}')
```

Insights
Substitutes (Sub):

Defenders (Df), Midfielders (Mf), and Forwards (Fw):

Goalkeepers (Gk):

Conclusion
The density plot provides a clear visual representation of how player ratings are distributed across different positions. It highlights that:

Substitutes have a more consistent and lower rating.
Goalkeepers have a broader and more even distribution of ratings.
Defenders, midfielders, and forwards have overlapping distributions but with different concentration points and spreads.
This type of visualization helps in understanding the general performance levels associated with each position, which can be useful for team formation, player assessment, and strategic planning.

## checking for outliers in each of the dataframes

```python
#data to check outlier for goalkeepers
# Select columns with data types int or float
goalkeepers_outlier_data = goalkeepers_df.select_dtypes(include=['int64', 'float64'])

# Check the shape of the resulting DataFrame
goalkeepers_outlier_data.shape
```

```python
from sklearn.cluster import DBSCAN

# Range of eps values to test
eps_values = [500, 1000, 5000, 10000, 20000]

for eps in eps_values:
    outlier_detection = DBSCAN(min_samples=10, eps=eps)
    clusters = outlier_detection.fit_predict(goalkeepers_outlier_data)
    TotalOutliers = list(clusters).count(-1)
    print(f"Total number of outliers identified with eps={eps}: {TotalOutliers}")
```

```python
#import the implementation of this algorihm from sklearn
from sklearn.ensemble import IsolationForest

#Use the algorithm for outlier detection, then use it to predict each point
#Any point labelled as -1 is an outlier
clf = IsolationForest(max_samples=3158, random_state = 2, contamination= 0.01)
preds = clf.fit_predict(goalkeepers_outlier_data)
#print(preds)
totalOutliers=0
for pred in preds:
    if pred == -1:
        totalOutliers=totalOutliers+1
print("Total number of outliers identified is: ",totalOutliers)
```

```python
from sklearn.neighbors import LocalOutlierFactor
np.random.seed(42)

# fit the model for outlier detection (default)
clf = LocalOutlierFactor(n_neighbors=25, contamination=0.01)

# use fit_predict to compute the predicted labels of the training samples
# (when LOF is used for outlier detection, the estimator has no predict,
# decision_function and score_samples methods).
y_pred = clf.fit_predict(goalkeepers_outlier_data)

#print ("Predicted outliers: \n",y_pred)

#Calculate number of outliers
outlierN = len([i for i in y_pred if i == -1])
print ("Number of predicted outliers:",outlierN)


#Let us see graphically how the outliers are identified
#Use outlier scores calculated by the algorithm to plot a circle around the outlier proportional to score.

X_scores = clf.negative_outlier_factor_

plt.figure(figsize=(12,8))
plt.title("Local Outlier Factor (LOF)")

#Convert the df to an array for the purpose of the scatterplot
Xol = goalkeepers_outlier_data.to_numpy()

#Use scatter 
plt.scatter(Xol[:, 0], Xol[:, 1], color='k', s=3, label='Data points')

# plot circles with radius proportional to the outlier scores
radius = (X_scores.max() - X_scores) / (X_scores.max() - X_scores.min())
plt.scatter(Xol[:, 0], Xol[:, 1], s=500 * radius, edgecolors='r',
            facecolors='none', label='Outlier scores')
plt.axis('tight')
#plt.xlim((0, 14))
#plt.ylim((0, 6))
plt.xlabel("N. Outliers: %d" % (outlierN))
legend = plt.legend(loc='upper left')
legend.legendHandles[0]._sizes = [12]
legend.legendHandles[1]._sizes = [20]
plt.show()
```

### Outlier Detection Method
I applied three distinct outlier detection methodsDBSCAN, Isolation Forest, and Local Outlier Factor (LOF)to understand the distribution of anomalies in our dataset. DBSCAN uses min_samples=10 and eps=10 to define dense clusters, effectively identifying outliers in sparse regions. Isolation Forest, employing max_samples=31258, contamination=0.01, and random_state=2, isolates anomalies through random feature selection and splitting, with the contamination rate adjusting sensitivity. Local Outlier Factor (LOF) utilizes n_neighbors=25 and contamination=0.01 to measure local density deviations, flagging significantly less dense points as outliers. The aim of this analysis is not to remove all detected outliers but to understand how each method characterizes them. This knowledge is crucial as our model needs to be robust enough to handle outliers effectively in real-world scenarios. Using various outlier detection methods allows for a comprehensive assessment of anomalies. This diversity in approaches ensures that we capture a broad spectrum of outliers, enhancing the robustness of our data analysis. Each detection method provides a unique perspective on identifying outliers, and their combined application offers a thorough exploration of potential data irregularities. This approach not only helps in fine-tuning our analysis but also in building models that are resilient in diverse conditions.
